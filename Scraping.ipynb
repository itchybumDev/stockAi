```python
from urllib.request import Request, urlopen

from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from dateutil.parser import parse

from bs4 import BeautifulSoup

def tableDataText(table):    
    """Parses a html segment started with tag <table> followed 
    by multiple <tr> (table rows) and inner <td> (table data) tags. 
    It returns a list of rows with inner columns. 
    Accepts only one <th> (table header/data) in the first row.
    """
    def rowgetDataText(tr, coltag='td'): # td (data) or th (header)
        td_text = [td.get_text(strip=True) for td in tr.find_all(coltag)]  
        td_text.append(tr.find('a', { 'class' : 'tab-link-news' })['href'])
        return td_text

    rows = []
    trs = table.find_all('tr')
    date = ""
    for tr in trs: # for every table row
#         datas = rowgetDataText(tr, 'td') 

        row_data = rowgetDataText(tr, 'td')
        if len(row_data[0]) > 8:
            date = row_data[0].split(' ')[0]
        else :
            row_data[0] = date + ' ' + row_data[0]
        rows.append(row_data) # data row       

    return rows

ans = {}
for i in range(1989, 1987, -1):
    url = "https://www.irishlottery.com/archive-{}".format(i)
    # response = urllib.request.urlopen(url)

    url_request = Request(url, headers={"User-Agent": "Mozilla/5.0"})
    webpage = urlopen(url_request).read()
    soup = BeautifulSoup(webpage, 'html.parser')
    htmltable = soup.find('table', { 'class' : 'table lotto archiveTable' })        
    #     return tableDataText(htmltable)
    trs = htmltable.find_all('tr')

    for tr in trs:
        if (tr.find('th') is None or tr.find('td') is None):
            continue
        date = parse(tr.find('th').get_text(strip=True)).date()
        lis = tr.find('td').find_all('li')
        numbs = []
        for li in lis:
            numbs.append(li.get_text(strip=True))
        ans[date] = numbs
        print([date, numbs])
   
```
